# PAM - Produção Agrícola Municipal

## Café

### Área x UF

```python
pam_cafe = pd.read_csv('../data/pam/output/cafe/pam_cafe.csv')
```

```python
geodata_uf = gpd.read_file('../data/gis-dataset-brasil/uf/shapefile/uf.shp')
```

```python
geodata_uf.head()
```

```python
pam_cafe.head()
```

```python
pam_cafe_uf = pam_cafe.groupby(['UF', 'ANO']).aggregate(sum).reset_index()[['UF', 'ANO', 'AREA_PLANTADA', 'QUANTIDADE_PRODUZIDA', 'RENDIMENTO_MED_PRODUCAO', 'VALOR_PRODUCAO']]
pam_cafe_uf = pam_cafe_uf[pam_cafe_uf['ANO'] >= 1990]
pam_cafe_uf.rename(columns={'UF':'NOME_UF'}, inplace=True)
pam_cafe_geo_uf = geodata_uf.merge(pam_cafe_uf, on='NOME_UF')
```

```python
pam_cafe_geo_uf
```

```python
pam_cafe_geo_uf.to_file("../data/pam/output/cafe/pam_cafe_geo_uf.geojson", driver='GeoJSON')
```

### Área x Município

```python
geodata_mun = gpd.read_file('../data/gis-dataset-brasil/municipio/shapefile/Munic.shp')
```

```python
pam_cafe.head()
```

```python
geodata_mun.head()
```

```python
pam_cafe.rename(columns={'MUNICIPIO':'NOME', 'CODIGO_MUNICIPIO':'GEOCODIGO'}, inplace=True)
pam_cafe_geo_mun = geodata_mun.merge(pam_cafe, on='GEOCODIGO')
pam_cafe_geo_mun = pam_cafe_geo_mun.drop(['UF_y', 'NOME_y', 'CODIGO_UF'], axis=1).rename(columns={'NOME_x':'MUNICIPIO', 'UF_x':'UF'})
```

```python
pam_cafe_geo_mun.head()
```

```python
ufs = pam_cafe_geo_mun['UF'].unique().tolist()
```

```python
for uf in ufs:
    df = pam_cafe_geo_mun[pam_cafe_geo_mun['UF'] == uf]
    df.to_file(f"../data/pam/output/cafe/pam_geo_ufs/pam_cafe_geo_{uf}.geojson", driver='GeoJSON')
```

# Café - Áreas de Origem

```python
cafe_areas_origem = pd.read_csv("../data/regioes_origem_cafe.csv", sep = ',', encoding='UTF-8')
```

```python
cafe_areas_origem = cafe_areas_origem.set_axis(['REGIAO_ORIGEM', 'NOME', 'UF'], axis=1)
cafe_areas_origem = cafe_areas_origem.sort_values(['UF', 'NOME']).reset_index().drop('index', axis=1)
cafe_areas_origem.head()
```

```python
geodata_mun.head()
```

```python
cafe_areas_origem.replace("São João d`Aliança", "São João d'Aliança", inplace=True)
cafe_areas_origem.replace("Olhos-d`Água", "Olhos-d'Água", inplace=True)
cafe_areas_origem.replace("Brazópolis", "Brasópolis", inplace=True)
cafe_areas_origem.replace("Pingo-d`Água", "Pingo-d'Água", inplace=True)
cafe_areas_origem.replace("Diamante d`Oeste", "Diamante D'Oeste", inplace=True)
cafe_areas_origem.replace("Rancho Alegre d`Oeste", "Rancho Alegre D'Oeste", inplace=True)
cafe_areas_origem.replace("Trajano de Moraes", "Trajano de Morais", inplace=True)
cafe_areas_origem.replace("Alta Floresta d`Oeste", "Alta Floresta D'Oeste", inplace=True)
cafe_areas_origem.replace("Alvorada d`Oeste", "Alvorada D'Oeste", inplace=True)
cafe_areas_origem.replace("Espigão d`Oeste", "Espigão D'Oeste", inplace=True)
cafe_areas_origem.replace("Machadinho d`Oeste", "Machadinho D'Oeste", inplace=True)
cafe_areas_origem.replace("Nova Brasilândia d`Oeste", "Nova Brasilândia D'Oeste", inplace=True)
cafe_areas_origem.replace("Santa Luzia d`Oeste", "Santa Luzia D'Oeste", inplace=True)
cafe_areas_origem.replace("São Felipe d`Oeste", "São Felipe D'Oeste", inplace=True)
cafe_areas_origem.replace("Mogi Mirim", "Moji Mirim", inplace=True)
cafe_areas_origem.replace("Itapajé", "Itapagé", inplace=True)
```

```python
gd_cafe_ao_acre = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'AC'], on=['NOME', 'UF'], how='right')
gd_cafe_ao_bahia = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'BA'], on=['NOME', 'UF'], how='right')
gd_cafe_ao_distritof = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'DF'], on=['NOME'], how='right')
gd_cafe_ao_espirito_s = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'ES'], on=['NOME'], how='right')
gd_cafe_ao_goias = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'GO'], on=['NOME', 'UF'], how='right')
gd_cafe_ao_minasgerais = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'MG'], on=['NOME', 'UF'], how='right')
gd_cafe_ao_matogrosso = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'MT'], on=['NOME'], how='right')
gd_cafe_ao_pernambuco = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'PE'], on=['NOME', 'UF'], how='right')
gd_cafe_ao_parana = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'PR'], on=['NOME', 'UF'], how='right')
gd_cafe_ao_riodejaneiro = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'RJ'], on=['NOME', 'UF'], how='right')
gd_cafe_ao_roraima = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'RO'], on=['NOME'], how='right')
gd_cafe_ao_saopaulo = geodata_mun.merge(cafe_areas_origem[cafe_areas_origem['UF'] == 'SP'], on=['NOME'], how='right')
```

```python
geodata_cafe_areas_origem = geodata_mun.merge(cafe_areas_origem, on=['NOME', 'UF'], how='right')
geodata_cafe_areas_origem[geodata_cafe_areas_origem['ID_UF'].isna() == True]
```

```python
geodata_cafe_areas_origem.head()
```

```python
geodata_cafe_areas_origem.to_file("../data/pam/output/cafe/pam_cafe_geo_areas_origem_municipal.geojson", driver='GeoJSON')
```
